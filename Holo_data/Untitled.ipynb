```python
def f2e(f):
    x,y,z,w = f
    rol = np.arctan2(2*(w*x + y*z), 1 - 2*(x**2 + y**2))
    yaw = np.arcsin(2 * (w*y - z*x))
    pit = np.arctan2(2*(w*z + x*y), 1 - 2*(y**2 + z**2))
    return [rol, yaw, pit]

ori = ego.ori.apply(f2e)
```
